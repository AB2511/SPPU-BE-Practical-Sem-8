Implementation of Tokenization, Stemming and Lemmatization using NLTK

### Problem Statement:

Perform tokenization (Whitespace, Punctuation-based, Treebank, Tweet, MWE) using NLTK library. Apply Porter Stemmer and Snowball Stemmer for stemming and use any technique for lemmatization.

```python
# Install (run only once if needed)
# !pip install nltk

import nltk
from nltk.tokenize import word_tokenize, wordpunct_tokenize, TweetTokenizer
from nltk.tokenize.treebank import TreebankWordTokenizer
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('wordnet')
```

```python
sentence = "Today we would learn about tokenization. Are you all ready?"
print("Original Sentence:\n", sentence)
```

## TOKENIZATION

```python
whitespace_tokens = sentence.split()
print("Whitespace Tokenization:\n", whitespace_tokens)
```

```python
punct_tokens = wordpunct_tokenize(sentence)
print("Punctuation-based Tokenization:\n", punct_tokens)
```

```python
treebank = TreebankWordTokenizer()
treebank_tokens = treebank.tokenize(sentence)
print("Treebank Tokenization:\n", treebank_tokens)
```

```python
tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(sentence)
print("Tweet Tokenization:\n", tweet_tokens)
```

## STEMMING

```python
words = ['run', 'runner', 'running', 'ran', 'runs', 'easily', 'fairly']
print("Original Words:\n", words)
```

```python
porter = PorterStemmer()

print("Porter Stemmer Output:")
for word in words:
    print(f"{word} → {porter.stem(word)}")
```

```python
snowball = SnowballStemmer("english")

print("Snowball Stemmer Output:")
for word in words:
    print(f"{word} → {snowball.stem(word)}")
```

## LEMMATIZATION

```python
lemmatizer = WordNetLemmatizer()

print("Lemmatization Output:")
for word in words:
    print(f"{word} → {lemmatizer.lemmatize(word, pos='v')}")
```

```python
sentence2 = "I saw eighteen mice running quickly"

tokens = word_tokenize(sentence2)

lemmatized = [lemmatizer.lemmatize(word, pos='v') for word in tokens]

print("Original Sentence:", sentence2)
print("Lemmatized Sentence:", " ".join(lemmatized))
```

## MWE (Multi-Word Expression)

```python
from nltk.tokenize import MWETokenizer

mwe = MWETokenizer()
mwe.add_mwe(('New', 'York'))

tokens = mwe.tokenize(sentence.split())
print("MWE Tokenization:", tokens)
```
